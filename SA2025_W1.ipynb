### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame
# Step 1: Group by origin and model_year, then calculate average mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Step 2: Multi-line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', marker='o')

plt.title('Evolution of Average MPG Over the Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.legend(title='Origin')
plt.grid(True)
plt.tight_layout()
plt.show()

```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
import pandas as pd

# Use the correct column name — either 'car_name' or 'name'
# Adjust below if needed
car_col = 'name' if 'name' in df.columns else 'car_name'

# Step 1: Filter car names that appear in more than one model year
multi_year_cars = df.groupby(car_col)['model_year'].nunique()
multi_year_cars = multi_year_cars[multi_year_cars > 1].index

# Step 2: Compute std(mpg) for these cars
mpg_std = df[df[car_col].isin(multi_year_cars)].groupby(car_col)['mpg'].std()

# Step 3: Filter for std(mpg) < 1.0
consistent_cars = mpg_std[mpg_std < 1.0].index

# Step 4: Prepare final report
result = (
    df[df[car_col].isin(consistent_cars)]
    .groupby(car_col)
    .agg(
        appearances=('model_year', 'nunique'),
        avg_mpg=('mpg', 'mean')
    )
    .reset_index()
)

# Step 5: BONUS — sort by appearances desc, then avg_mpg desc
result = result.sort_values(by=['appearances', 'avg_mpg'], ascending=[False, False])

# Display result
print(result)
```
